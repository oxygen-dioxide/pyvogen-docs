# 从music21合成歌声
Vogen提供了music21 api，支持将music21的单轨与多轨对象，包括stream, measure, part和score对象导入到vogen进行合成。

```python
import music21
import vogen
from vogen import synth
```

首先，创建music21对象。

这里使用music21的tinyNotation创建对象。关于tinyNotation语法，参见[music21文档](https://web.mit.edu/music21/doc/usersGuide/usersGuide_16_tinyNotation.html)

```python
#乐谱
stream=music21.converter.parse("""tinyNotation: 4/4 r1
G4 E8 G c4. r8
A4 c G4. r8
G4 C8 D E4 D8 C
D2. r4
G4 E8 G c4. B8
A4 c G4. r8
G4 D8 E F4. BB8
C2. r4""", makeNotation=False)
#显示五线谱
stream.show()
```

```python
#歌词
lyrics="长亭-外，古道边，芳草-碧连-天。晚风-拂柳笛声残，夕阳-山外山。".replace("，","").replace("。","")
#灌入歌词
for n, lyric in zip(stream.notes, lyrics):
    n.lyric = lyric
stream.show()
```

然后导入vogen进行合成

```python
#将music21 stream对象转为vogen VogFile对象
vf=vogen.toVogFile(stream).lyrictorom()
vf
```

```python
#保存文件
vf.save("songbie.vog")
```

```python
#合成
vf.play()
```
